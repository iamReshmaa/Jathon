```python
# 1. Find the maximum and min element in an array
def minn(arr):
  ans = arr[0]
  for i in range(1,len(arr)):
    if ans>arr[i]:
      ans = arr[i]
  return ans
def maxx(arr):
  ans = arr[0]
  for i in range(1,len(arr)):
    if ans<arr[i]:
      ans = arr[i]
  return ans
arr = []
arr = list(map(int, input("Enter numbers: ").split()))
print(minn(arr))
print(maxx(arr))
```

```python
# 2. Find the second largest element in an array
def secmaxx(arr):
  if arr[0]>arr[1]:
    larg = arr[0]
    sec = arr[1]
  else:
    larg = arr[1]
    sec = arr[0]
  for i in range(2,len(arr)):
    if arr[i] > larg:
      sec = larg
      larg = arr[i]
    elif arr[i]>sec and arr[i] != larg:
      sec = arr[i]
  return sec



arr=list(map(int,input().split()))
print(secmaxx(arr))

```

```python
# 3. Rotate an array by k positions
def rotate(arr,k):
  ans = arr
  for i in range(0,len(arr)):
    ans[i] = arr[k]
    k = k+1
  return ans
arr = list(map(int,input().split()))
k = int(input())
rotate(arr,k)
```

```python
#T1. Print all divisors of given number
def div(n):
  arr = []
  for i in range(1,n+1):
    if n%i == 0:
      arr.append(i)
  return arr
n = int(input())
div(n)
```
